```python
import numpy as np
import matplotlib.pyplot as plt

MAE_dict = {
    8: 0.105231,
    16: 0.049630,
    32: 0.058231,
    64: 0.049811,
    80: 0.051053
}
MSE_dict = {
    8: 0.021593,
    16: 0.005434,
    32: 0.007691,
    64: 0.006829,
    80: 0.006664
}

hidden_sizes = list(MAE_dict.keys())
mae_val = list(MAE_dict.values())
mse_val = list(MSE_dict.values())

plt.figure(figsize=(8, 6))
plt.plot(hidden_sizes, mae_val, marker='s', color='blue', label='MAE Loss')
plt.plot(hidden_sizes, mse_val, marker='^', color='red', label='MSE Loss')
plt.xlabel('Hidden Size')
plt.ylabel('Loss')
plt.legend()
plt.title('LSTM模型不同Hidden Size的Loss值', fontproperties='SimHei')
# 设置 x 轴刻度为 hidden_size
plt.xticks(hidden_sizes)
plt.grid(True)

plt.show()
```
